- 实现一个简单的具有加减法功能的计算器

    - GUI
    - 给每个控件配置相应的事件
    - 写逻辑代码


```python
import tkinter

from tkinter import *

root = Tk()
# 定义面板大小
root.geometry('380x480')
root.title('军仔计算器')

# 定义面板，bg代表背景background，#dddddd是十六进制表示颜色的一个串
frame_show = Frame(width=400, height=200, bg='#000000')

# 定义顶部区域
sv = StringVar()

sv.set('0')# 初始化

# anchor，定义控件的锚点，e表示east，东边，也就是右。
# font代表字体（字体，字号，字形）
show_label = Label(frame_show, textvariable=sv, bg='yellow', width=20, height=3, font=('黑体',20,'bold'), \
                  justify=LEFT, anchor='e')
# label创建完了，得放到一个地方去
show_label.pack(padx=10, pady=10)
frame_show.pack()

def delete():
    global num1
    global num2
    global num1_
    global num2_
    global operator
    if not operator:
        sv.set(num1_)
        
    elif operator:
        sv.set(num1 + operator + num2_)   
    # 这步只实现了删除一位，并且没写删除操作符的代码。如果想实现多删，应该从一开始就建立列表来储存输入信息
    # 那么这里的delete功能只要执行一个pop（）就好了，这里懒得写了。
    pass

def clear():
    global num1
    global num2
    global operator
    
    sv.set('')
    
    num1 = ''
    num2 = ''
    operator = ''

def fan():
    global num1
    global num2
    global operator
    if not operator:
        num1 = '-' + num1
        sv.set(num1)
        
    else:
        num2 = '-' + num2
        sv.set(num1 + operator + num2)   


# 按键区域
frame_bord = Frame(width=400, height=350, bg='#ffffff')

# 定义按钮
b_del = Button(frame_bord, text='←', width=10, height=2, command=delete, anchor='center').grid(row=0, column=0)
b_clear = Button(frame_bord, text='CE', width=10, height=2, command=clear).grid(row=0, column=1)
b_fan = Button(frame_bord, text='±', width=10, height=2, command=fan).grid(row=0, column=2)

'''按下数字就显示，但是按完之后，要转为整型.考虑几种情况
1 按下数字
2 按下操作符
3 只考虑两个数的操作，不考虑其他复杂情况'''
num1 = ''
num2 = ''
operator = ''
num1_ = ''
num2_ = ''


def change(num):
    global num1
    global num2
    global num1_
    global num2_
    global operator
    '''
    按下数字在这里实现，有两种情况，数字属于第一个操作数，或者第二个
    判断是否属于第一个，可以通过operator
    '''
    if not operator:
        num1_ = num1
        num1 = num1 + num
        # 如果是第一个操作数，则只显示第一个操作数
        sv.set(num1)
        
    else:
        num2_ = num2
        num2 = num2 + num
        # 如果是第二个操作数，则显示全部
        sv.set(num1 + operator + num2)            
    
def operation(op):
#     global op
    global operator
    
    if op in ['+','-','x','/']:
        operator = op
        sv.set(num1 + operator)
    if op in ['=','o']: 
        if operator == '+':
            result = int(num1) + int(num2)
        if operator == '-':
            result = int(num1) - int(num2)
        if operator == 'x':
            result = int(num1) * int(num2)
        if operator == '/':
            result = int(num1) / int(num2)
    
        sv.set(num1 + operator + num2 + op + str(result)) # 我草了，就少打个括号，调试20分钟，太可怕了。

# 写数字按钮
b_1 = Button(frame_bord, text='1', width=10, height=4, command=lambda:change('1')).grid(row=1,column=0)
b_2 = Button(frame_bord, text='2', width=10, height=4, command=lambda:change('2')).grid(row=1,column=1)
b_3 = Button(frame_bord, text='3', width=10, height=4, command=lambda:change('3')).grid(row=1,column=2)
b_4 = Button(frame_bord, text='4', width=10, height=4, command=lambda:change('4')).grid(row=2,column=0)
b_5 = Button(frame_bord, text='5', width=10, height=4, command=lambda:change('5')).grid(row=2,column=1)
b_6 = Button(frame_bord, text='6', width=10, height=4, command=lambda:change('6')).grid(row=2,column=2)
b_7 = Button(frame_bord, text='7', width=10, height=4, command=lambda:change('7')).grid(row=3,column=0)
b_8 = Button(frame_bord, text='8', width=10, height=4, command=lambda:change('8')).grid(row=3,column=1)
b_9 = Button(frame_bord, text='9', width=10, height=4, command=lambda:change('9')).grid(row=3,column=2)
b_0 = Button(frame_bord, text='0', width=10, height=4, command=lambda:change('0')).grid(row=4,column=0)

# 写操作符按钮
b_plus = Button(frame_bord, text='+', width=10, height=4, command=lambda:operation('+')).grid(row=4,column=1)
b_sub  = Button(frame_bord, text='-', width=10, height=4, command=lambda:operation('-')).grid(row=4,column=2)
b_xxx  = Button(frame_bord, text='x', width=10, height=4, command=lambda:operation('x')).grid(row=5,column=0)
b_chu  = Button(frame_bord, text='/', width=10, height=4, command=lambda:operation('/')).grid(row=5,column=1)
b_deng = Button(frame_bord, text='=', width=10, height=4, command=lambda:operation('=')).grid(row=5,column=2)

# 导入全部按钮
frame_bord.pack(padx=10, pady=10)

root.mainloop()
```
