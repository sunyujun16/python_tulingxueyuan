## 参数详解
- 参数分类
    - 普通参数
    - 默认参数
    - 关键字参数
    - 收集参数
    
- 普通参数
    - 最普通的参数
    - 直接定义变量名
    - 调用时候把值放入指定变量名
    
- 默认参数
    - 形参带着默认值，你不输入它也不报错，而是用默认值
                
                def fuc_name(p1=v1,p2=v2...):
                
                # 调用1
                fuc_name()
                
                #调用2
                fuc_name(value1,value2...)
                
- 关键字参数
    - 语法
            
            def func(p1=v1,p2=v2,...):
                
            调用函数：
            func(p1=value1,p2=value2...)
            
    - 比较麻烦，但也有好处，按位置对照的话容易出错
    - 使用关键字参数，可以不考虑参数位置
    
- 收集参数
    - 把没有位置，不能和定义时的参数位置相对应的参数，放入一个特定的数据结构中
    - 语法
    
            def func(*args):
                func_body
                按照list的访问方式，访问args得到传入的参数
            
            调用：
            func(p1,p2,p3,...)
    - 参数名不一定非要是args，但是大家都这么写
    - args之前需要加*，表明他是一个收集参数
    - 收集参数可以和其他参数共存

## 收集参数混合调用的顺序问题
- 收集参数，关键字参数，普通参数可以混合使用
- 使用规则就是，普通参数和关键字参数优先处理
- 定义的时候，一般找普通参数，关键字参数，收集参数tuple，收集参数dict


```python
# 举个栗子
# 收集参数混合调用
# 注意定义和调用时，输入形参、实参的先后顺序。

def stu(name, age, *args, hobby='None', **kwargs):
    print('I am {0},I\'m {1} years old'.format(name,age))
    if hobby == 'None':
        print('没什么爱好')
    else:
        print('爱好是{0}'.format(hobby))
    
    print('* '*15)

    for i in args:
        print(i)
    
    print('= '*15)

    for k,v in kwargs.items():
        print(k,',',v)

# diaoyong
name = '孙宇军'
age = 26

# 注意下面的参数怎么传入函数的
stu(name,age,'收集参数-试试看','收集参数-再试试看','收集参数-一定要试试看',gender='female',hobby='刘奕含',hobby2 = "睡觉")
```

# 递归
- 函数直接或者间接调用自己
- 优点：简洁，理解容易
- 缺点：对递归深度有限制，资源消耗较大
- Python对递归深度有限制，超过限制报错
- 递归一定要注意写好结束条件

```python
# 递归调用深度限制

x = 0
print(x)


def func():
    global x 
    x += 1
    print(x,end=',')
    
    func()
    
# 调用函数
func()
# 1965次时被终止
```

```python
# 斐波那契数列
# 一列数字，1,1，sum前两个

def feb(n):
    
    if n == 1:
        return 1
    
    if n == 2:
        return 1
    
    else:
        return feb(n-1) + feb(n-2)

feb(5)

```

## 6.1 封装
- 封装就是对对象的成员进行访问限制
- 封装的三个级别
    - 公开：public
    - 受保护的：protected
    - 私有的：private
    - 这三个p，这个3p，不是关键字
- 判别对象的位置
    - 对象内部
    - 对象外部
    - 子类中
- 私有
    - 私有是最高级别的封装，只能在当前类或对象中访问
    - 在成员前面添加两个下划线
    
                class Person():
                    # name 是公用的成员
                    name = 'sun'
                    # __age就是私有成员
                    __age = 18
                    
    - Python的私有并不是真私有，是一种称为name mangling的改名策略，
        可以使用对象__classname__.attributename

- 受保护的封装
    - 受保护封装是将对象成员进行一定级别的封装，然后在类或者子类中都可以访问，但是在外部不可以
    - 封住哪个方法，在成员名字前加一个下划线即可
    
            class Person():
                _age = 22
- 公开的，公共的public
    - 公共的封装实际对成员没有任何操作。
    
## 6.2 继承
- 不重复造轮子，继承就是一个类可以获得另一个类的成员属性和方法
- 作用，减少代码，增加代码的复用功能，同时可以设置类与类直接的关系。
- 继承与被继承的概念
    - 父类、基类、超类 basecalss
    - 子类、派生类 subclass
    - 他们之间是 is a 的关系
- 继承的语法，略
- 继承的特征
    - 所有类都继承自object，普天之下莫非王土，率土之滨莫非王臣
    - 子类一单继承，则可以使用除private成员外的所有内容
    - 子类继承后并不会重新赋值，而是引用父类的现有成员
    - 子类中有与父类同名的成员，优先使用子类(LEGB法则)
    - 子类如果想扩充父类的方法，可以在定义新方法的同时访问父类成员来进行代码重用，
    - 可以使用‘父类名.父类成员’的方式来调用父类成员，也可以使用‘super().父类成员’的格式来调用
- 继承变量函数的查找顺序问题

```python
class MappingMixin:
    def __init__(self):
        self.key = None
    
    def __getitem__(self, key):
        a = self.__dict__.get(key)
        self.key = key
        print('滚犊子')
        return a

    def __setitem__(self, key, value):
        self.__dict__[key] = value
        print("去你妈的")


class Person(MappingMixin):
    def __init__(self, name, gender, age):
        self.name = name
        self.gender = gender
        self.age = age
        
    def __setitem__(self, key, value):
#         self.__dict__[key] = value
        print("去你妈的")
        
p = Person("小陈", "男", 18)
print(p.name)
print(p.__dict__)
print(p['name'])  # 注意__getitem__是在这时候自动执行的
print(p['age'])
p['age'] = 88 #setitem触发
# p.age = 88 # setitem没有触发,说明要采用字典方式修改才会触发
print(p['age'])
print(p.__dict__)
```

```python

class A():
    def __init__(self):
        self.name = 'haha'
        self.age = 18
       
    # 此功能，是对类变量进行读取操作的时候执行的函数
    def fget(self):
        print('读取的过程')
        return self.fakename2
    
    # 对变量进行写操作时触发
    def fset(self,name2):
        print('写入的过程')
        self.fakename2 = name2 + '刘奕含'
        
    # 模拟删除的时候进行的操作
    def fdel(self):
        pass
    
    
    # property 的四个函数顺序是固定的
    # 第一个参数代表读取函数，第二个代表写入，第三个删除，第四个xx
    name2 = property(fget,fset,fdel,'这是property案例')
    
a = A()

print(a.name)
a.name = '孙宇军' # 这里并没有调用写入函数，因为函数定义的是name2

print('='*30)
a.name2 = '孙宇军' # 这里定义name2，这才触发写入函数

print('='*30)
print(a.name2) # 这里读取了name2的值，触发fget，但由于之前触发了写入函数，所以这里不是单独的打印孙宇军
# 实际上是打印的a.fakename2
```

```python
# 抽象类

import abc

# 声明一个类并且指定当前类的元类
class Human(metaclass=abc.ABCMeta):

    # 定义一个抽象方法
    @abc.abstractmethod
    def smoking(self):
        pass
    
    # 定义类抽象方法
    @abc.abstractclassmethod
    def drink(cls):
        pass
    
    # 定义静态抽象方法
    @abc.abstractstaticmethod
    def fuck():
        pass
    
    def sleep(self):
        print('sleeping......')
    
```

```python
# 元类演示
# 元类一般以MetaClass结尾
# 元类写法是固定的，必须继承自type

class TuLingMetaClass(type):
    # 注意下一行写法固定，不能自由发挥
    def __new__(cls,name,bases,attrs):
        # 然后写自己的业务处理
        print('介四个元类')
        attrs['id'] = 9527
        attrs['addr'] = '公主坟'
        
        return type.__new__(cls,name,bases,attrs)
    
# 元类定义完就可以使用，使用注意写法

# 下面这句会对元类中的print进行执行，其实可以理解为元类的构造函数，和’实例化‘一样，# 说的不对,是实例化元类时候自动调用new函数而已
# 我们在下面这句进行了’类化‘，于是乎运行了上面__new__函数的内容，所以一个类就这样诞生了。
class Teacher(object,metaclass=TuLingMetaClass): # （父类，引用元类）
    pass
print('=====')
t = Teacher()

print(t.__dict__) # 温习一下，l类属性不会因为实例化就传给实例（instance）
# print(Teacher.__dict__) # 但是会传给子类哦

print('属性id为：',t.id)
print('属性地址为：',t.addr)
```

```python
# 那么怎么能让赋值有用呢，想必是要在描述符的函数内下功夫

class Descriptors():
 
    def __init__(self, key, value_type):
        self.key = key
        self.value_type = value_type
 
    def __get__(self, instance, owner):
        print("执行Descriptors的get")
        return instance.__dict__[self.key]
 
    def __set__(self, instance, value):
        print("执行Descriptors的set")
        if not isinstance(value, self.value_type):
            raise TypeError("参数%s必须为%s"%(self.key, self.value_type))
        instance.__dict__[self.key] = value 
 
    def __delete__(self, instance):
        print("执行Descriptors的delete")
        instance.__dict__.pop(self.key) 
        
        
class Person():
 
    name = Descriptors("芳名", str)
    age = Descriptors("芳龄", int) 
    # 这两句使得name和age和上面有了关联，在对他俩进行调用或者设置时，就会自动跳转上面，而忽略下面。
 
    def __init__(self, name0, age0):
        self.name = name0
        self.age = age0
 
 
person = Person("奕含", 26)
print(person.name)
# 这里打出了奕含，要知道，尽管跳转到上面的描述符，但是描述符仍然是作为一个整体，而在Person class内部运行的，这就是为什么明明是
# set函数对参数进行了处理，结果却会反应到person的属性当中。其对应关系应该是：self无意义，instance对person，value对参数
# 而name = Descriptors("name", str)这个对应了Deacriptors的__init__函数，’name‘为key，str为value_type。
print('='*50)

print(person.__dict__) # 
print(Person.__dict__) # 注意value的具体值，奕含，只是加到了实例的属性当中
print('='*50)

print(person.name) # 
print('='*50)

person.name = "jone" # 
print(person.__dict__)

# 到这儿基本就明白了，呼～可以午睡了

# 20201111复习纠正: name 是Descriptors的实例,这是关键点, 当打印person.name的时候
# 实际上是在打印__get__函数的返回值, 这些值来来回回都是在Descriptors内部赋值的.从下面的__dict__结果中也能看出这点
# 那么,描述符,实际上就是把属性设定成描述符类的实例这么个事儿.
# 果然温故而知新呐
```
