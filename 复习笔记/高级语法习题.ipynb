找出0-1000的回数: 从左向右和从右向左读都是一样的数, 例如1234321, 利用filter函数

```python
def eq(a, b):
    return a == b

def is_palindrome(n):
    s = str(n)
    for i in range(int(len(s)/2)):
        if s[i] != s[len(s)-1-i]:
            return False
        else:
            return True
        
output = filter(is_palindrome, range(0, 1000))
print(list(output))

```

编写一个程序,统计当前目录下每个文件类型的文件数
- 思路
    - 打开当前文件夹
    - 获取所有文件夹和文件名
    - 获取后缀
    - 形成键值对,用for循环叠加

```python
import os
def get_suffix_num():
    file_list = os.listdir(os.curdir)

    file_dict = {}
    for file in file_list:
        if os.path.isdir(file):
            file_dict.setdefault("文件夹", 0)
            file_dict["文件夹"] += 1
        else:
            suf = os.path.splitext(file)[1]
            if suf == "":
                suf = "无后缀"
            file_dict.setdefault(suf, 0)
            file_dict[suf] += 1

    for file in file_dict.keys():
        print("{}的数量有{}个".format(file, file_dict[file]))
        
get_suffix_num()
```

#### 装饰器
- 使用装饰器, 打印函数执行的时间

```python
import logging

DATE_FMT = "%Y-%m-%d %H:%M:%S" 
LOG_FORMAT = "%(asctime)s - %(levelname)s - %(message)s"
logging.basicConfig(format=LOG_FORMAT, level=logging.DEBUG, datefmt=DATE_FMT)


def log(text):
    def decorator(func):
        def wrapper(*arg, **kw):
#             print(text)
            logging.debug("执行了:{}".format(text))
            return func(*arg, **kw)
        
        return wrapper
        
    return decorator


@log('func1执行')
def func1(s):
    print("1的{}".format(s))
    return None


@log('func2执行')
def func2(s):
    print("2的{}".format(s))
    return None


func1("函数")
func2("函数")
```

#### 用logging的四大组件来实现日至的功能
- 打印出函数执行时间, 日志等级, 日志消息
- 用装饰器
- 不同的日志要记录不同等级的日志消息

```python
import logging

logger = logging.getLogger("my_logger")
logger.setLevel(logging.DEBUG)

# handler
# TimeRotationFileHandler 按日期划分日志
# RotationHandler 按文件大小划分日志

debug_handler = logging.FileHandler("1024debug.log")
debug_handler.setLevel(logging.DEBUG)
debug_handler.setFormatter(logging.Formatter("%(asctime)s - %(levelname)s - %(message)s -%(my_name)s[%(my_age)s] - years old"))

error_handler = logging.FileHandler("1024error.log")
error_handler.setLevel(logging.ERROR)
error_handler.setFormatter(logging.Formatter("%(asctime)s - %(levelname)s - %(message)s -%(my_name)s[%(my_age)s] - years old"))

logger.addHandler(debug_handler)
logger.addHandler(error_handler)


def my_log(name, age):
    def decorator(func):
        def wrapper(*arg, **kw):
            logger.debug("写点啥吧", extra={'my_name':name.capitalize(), 'my_age':age})
            logger.error("写点啥吧", extra={'my_name':name.capitalize(), 'my_age':age})
#             logger.debug(name) 报错,因为格式字符串里的my_name无法获得一个值.
#             logger.error(age)
            return func(*arg, **kw)
        return wrapper
    return decorator

@my_log('liuyihan', 26)
def test1():
    print("111")

@my_log('sunyujun', 26)
def test2():
    print('222')
    
test1()
test2()

# 注意重启jupyter解释器
```

#### 上面cell的调用顺序:
- test1() = mylog('liuyihan', 26) (test1) () = decorator(test1)() = wrapper()
- 而wrapper的返回值正是test1(), 只不过返回之前偷偷打印了日志

# 多线程

#### 使用多线程去播放两个列表, 一个是movie, 一个是music
- _treading
- threading

```python
import threading
import time

movie_list = ["Leon.mkv", "ForestGump.avi", "Devil.mp4", "nothing.rmvb"]
music_list = ["Adele.mp3", "Taylor.mp3"]

movie_format = ['mkv', 'avi', 'mp4']
music_format = ['mp3', 'ape']


def play(items):
    for item in items:
        if item.split(".")[1] in movie_format:
            print("开始播放视频%s"%item, time.ctime())
            time.sleep(1)
        elif item.split(".")[1] in music_format:
            print("开始播放音乐%s"%item)
            time.sleep(1)
        else:
            print("不能播放%s文件"%item)
            time.sleep(1)
        

def mul_thr():
    t1 = threading.Thread(target=play, args=(movie_list,))
    t2 = threading.Thread(target=play, args=(music_list,))
    t1.start()
    t2.start()
    t1.join()
    t2.join()
    
if __name__ == '__main__':
    mul_thr()
    
```

```python
import threading
import time

movie_list = ["Leon.mkv", "ForestGump.avi", "Devil.mp4", "nothing.rmvb"]
music_list = ["Adele.mp3", "Taylor.mp3"]

movie_format = ['mkv', 'avi', 'mp4']
music_format = ['mp3', 'ape']


def play(items):
    for item in items:
        if item.split(".")[1] in movie_format:
            print("开始播放视频%s"%item, time.ctime())
            time.sleep(1)
        elif item.split(".")[1] in music_format:
            print("开始播放音乐%s"%item)
            time.sleep(1)
        else:
            print("不能播放%s文件"%item)
            time.sleep(1)
        

class Play(threading.Thread):
    def __init__(self, play_list):
        super().__init__()
        self.play_list = play_list
        
    def run(self):
        play(self.play_list)

p1 = Play(movie_list)
p2 = Play(music_list)

p1.start()
p2.start()

    
```
